# Daily flower web scraping bot

### Importing libraries
#### *List of libraries used :* 
datetime is used to get the date of the day.\
bs4 is used for web scraping.\
requests is used for urls.\
re is used to locate integers in string lists.\
random is used to generate random numbers.\
PIL and io are used to display images in the python console.

```python
import numpy as np
from datetime import date
from bs4 import BeautifulSoup
import requests
import re
from random import randint
from PIL import Image
from io import BytesIO
```

### Step 1: Getting today's date

```python
[weekday,day,month,year] = [date.today().strftime(f"%{id}") for id in ["A","d","B","Y"]]
```

```python
day = int(day)
year = int(year)
```

### Step 2: Webscraping from the online flower database
Source : https://www.flower-db.com/en \
We will be looking for a flower which blooms in the current month, then selecting a consistent (but arbitrary) species within the list.

#### Step 2.1 : Create the daily url
We start by accessing the link to the first page of the month and identify how many pages exist in total.\
The url is then changed to incorporate the day, with the page number being equal to the day modulo the total length of pages to avoid erroneous urls.

```python
url = f"https://www.flower-db.com/en/search/seasons?q={month.lower()}"
soup = BeautifulSoup(requests.get(url).text,'html.parser')
```

```python
page_numbers = [num for num in re.findall("\d+", soup.get_text())]
page_count = str(page_numbers[-1])
```

```python
url = f"https://www.flower-db.com/en/search/seasons?q={month.lower()}&page={page_count}"
```

#### Step 2.2 : Choose a flower species of the day
We access the url obtained previously, extract the list of species and their relevant photo urls, then choose a random species from the available list.

```python
soup = BeautifulSoup(requests.get(url).text,"html.parser")
```

```python
daily_species = [a.attrs['alt'].title().lstrip() for a in soup(class_ = 'card-img-top') if 'alt' in a.attrs.keys()]
daily_species_urls = [a.attrs['src'] for a in soup(class_ = 'card-img-top') if 'alt' in a.attrs.keys()]
```

```python
rng_number = randint(0,len(daily_species)-1)
```

#### Step 2.3 : Save flower of the day details
We then save the details of the flower of the day.

```python
class flower_of_the_day():
    name = ''
    url = ''

flower_of_the_day.name = daily_species[rng_number]
flower_of_the_day.url = daily_species_urls[rng_number]
```

### Step 3: Displaying the flower of the day

```python
response = requests.get(str(flower_of_the_day.url))
img = Image.open(BytesIO(response.content))

print(f"Today is {weekday} {day} {month} {year}.")
print(f"The {month}-blooming flower of the day is the {flower_of_the_day.name}!")
display(img)
```
